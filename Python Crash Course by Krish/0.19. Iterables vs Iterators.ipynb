### **Iterables vs Iterators**
- **Iterables:** All the values are allocated in a memory location. And we can iterate all the values.
Example:
- List is an iterables

- **Iterators:** Not all the values are allocated in a memory. By using 'Next', the value will be saved in the memory and we can see the elements. With 'Next', only one value can be saved in the memory for a moment.

**Iterables**

```python
lst = [1,2,3,4,5,6,7]
for i in lst:
  print(i, end = " ")
```

**Change it to Iterators**

```python
lst1 = iter(lst)
print(lst1)
```

```python
next(lst1) #Run the code, the value will be changed every time until the list completed.
```

**For loop can also be used for iterators.**: For loop stop iterations by itself.

```python
for i in lst1:
  print(i)
```

### **Iterable to iterator in a single code**

```python
lsts = [1,2,3,4,5]
for i in lsts:
  print(i, end = " ")

lsts1 = iter(lsts)
for i in lsts1:
  print("\nIterators are: ", i)
```
