```python
my_str = "Grace Joydhar"
print(my_str.isalnum()) #if all characters are mixture of num and alphabets
print(my_str.isalpha()) #If all char are string are alphabets only
print(my_str.isdigit()) #if all chars are digit/number
print(my_str.istitle()) #if all the first char of a string is capital letter. Taht will be title
print(my_str.isupper()) #if all chars are uppercase
print(my_str.islower()) #if all chars are lowercase
print(my_str.isspace()) #if there is a blank space only
print(my_str.startswith("G")) #iTrue, f a string starts with G
print(my_str.endswith('R')) #True, if a string ends with r
```

```python
alpha = "Hello"
print("alpha:", alpha.isalpha())

digit = "123"
print("digit:",digit.isdigit())

alnum = "grace017"
print("alnum:", alnum.isalnum())

title = "My Website" #"My website" will be False
print("title:", title.istitle())

upper = "GRACE"
print("upper:", upper.isupper())

lower = "joydhar"
print("lower:", lower.islower())

space1 = " "
space2 = " g r a c e"
print("space1:", space1.isspace())
print("space2:", space2.isspace())

start_end = "Grace Joydhar"
print("starts with:", start_end.startswith("G"))
print("ends with:", start_end.endswith("r"))
```

**Append List:** *Add values in a list and add at the last.*
Make nested values.

```python
names = ['Jack', 'Mike', 'Jemy']
names.append('Grace')
names.append(['Jason', 'Middy']) #will make a nested loop
print(names)

print(names[2]) #print through index number

#Indexing Techniques
print(names[:2]) #will give all the elements before index 2
print(names[1:]) #will give all the elements from index 1 to last
print(names[-1]) #will print the last value from the list
```

**Insert:** *Add elements in a specific index*
Make nested values

```python
colors = ['red', 'black', 'white']
colors.insert(1, 'blue') #index number, value
colors.insert(4, ['perple', 'yellow'])
print(colors)
```

**Extend Method:** It will add values in the list. but *will not make any nested values.*

```python
list = [1,2,3,4,5]
list.extend([6,7])
print(list)
```

**Various Operations that we can perform in list**

```python
num = [5,2]
print(sum(num))
```

```python
num = [1,2,3]
print(num*5) #will print the list 5 times
```

```python
num = [1,2,3,4,5]
num.pop() #will remove the last digit
print(num) 

num.pop(1) #will remove the digit present in index 1
print(num)
```
