**1. Use Format Option (1st Way)**

```python
name = input("Enter your name: ")
age = input("Enter your age: ")
country = input("Enter your country name: ")

def greeting(name):
  return "Hello {}. Welcome to our page.".format(name)

showName = greeting(name)
print(showName)

def details(age, country):
  return "We think you are {} years old. And you are from {}".format(age, country)

show_Details = details(age, country)
print(show_Details)
```

**2. Use Format Option (2nd Way)** Write the names inside the {} bracket. This names are called placeholders.

```python
name = input("Enter your name: ")
age = input("Enter your age: ")
country = input("Enter your country name: ")

def greeting(name):
  return "Hello {name}. Welcome to our page.".format(name = name)

showName = greeting(name)
print(showName)

def details(age, country):
  return "We think you are {age} years old. And you are from {country}".format(age = age, country = country)

show_Details = details(age, country)
print(show_Details)
```
