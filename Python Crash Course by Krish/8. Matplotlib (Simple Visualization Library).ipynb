##**Matplotlib Tutorial**
Matplotlib is a ploting library for the python progrsmming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUL toolkits like Tkinter, wxPython, Qt orGTK+.

Some of the major pros of Matplotlib are:
- Generally easy to get started for simple plots
- Support for custom labels and text
- Great control of every element in a figure
- High quality output in many formats
- Very customizable in general

```python
#plotting using matplotlib
#plot scatter

plt.scatter(x, y, c='green', marker='*', linewidths=5)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Graph in 20") 
plt.show()
plt.savefig('Graph in 20.png') #Will save the file temporarily in the google colab storage

#Save the file in drive
from google.colab import drive
drive.mount('/content/gdrive')
plt.savefig('/content/gdrive/My Drive/Colab_Notebooks/Sample_data/Figures/Graph in 20.png')
```

## **Plot Graph**

```python
#plt plot
x = np.arange(0,10)
y = np.arange(11,21)

y = x*x

plt.plot(x,y, 'b*', linestyle = 'dashed', linewidth = 2, markersize = 8)
plt.xlabel("ABC")
plt.xlabel("X2 Axis")
plt.ylabel("Y2 Axis")
plt.title("2D Diagram")
plt.show()
```

**Subplots**

```python
#Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y, c='red')

plt.subplot(2,2,2)
plt.plot(x*y, c = 'green')

plt.subplot(2,2,3)
plt.plot(x*x, c = 'blue')

plt.subplot(2,2,4)
plt.plot(y*x, c = 'black')
```

**Complicated Operations**

```python
x = np.arange(1,11)
y = 3 * x + x * 10

plt.plot(x,y, c="purple")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Matplotlib Demo")
plt.show()
```

```python
#Compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4*np.pi, 0.1) # range is 0 to 4*np.pi. And the step is 0.1
y = np.sin(x)

plt.title("Sine Curve")
plt.plot(x, y)
plt.show()
```

**Subplots with sine and cosine**

```python
x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin, c = 'purple')
plt.title("Y_Sine")

plt.subplot(2,1,2)
plt.plot(x, y_cos, c = 'black')
plt.title("Y_Cosine")

plt.tight_layout() #to make space in between graphs.
plt.show()
```

## **Bar Graph**

```python
x1 = np.array([2,8,10])
y1 = np.array([11,16,9])

x2 = np.array([3,9,11])
y2 = np.array([6,15,7])

plt.bar(x1, y1, color = 'purple', align='center')

plt.bar(x2, y2, color = 'maroon')

plt.title("Bar Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

## **Histograms**
Helps to know the density of the given numbers.

```python
a = np.array([22,87,5,40,56,73,55,54,11,51,5,79,42,84])

plt.hist(a)
plt.title("Histogram Graph")
plt.show()
```

## **Box Plot**
- Helps to find out percentiles

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)] #std means standard deviation

#rectangular box plot
plt.boxplot(data, vert = True, patch_artist=True); #Patch_artist = False means, there will be no color. and Vert= False means the graph will be shown as horizontay.
```

## **Pie Chart**

```python
labels = 'Python', 'C++', 'Java', 'Ruby'
sizes = [235, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'skyblue', 'coral']
explode = (0.1,0,0,0) #Explode 1st slice

plt.pie(sizes, explode=explode, colors=colors, labels=labels, autopct = '%1.1f%%', shadow=True)

plt.legend(['Python', 'C++', 'Java', 'Ruby'], loc= 'center right')
plt.axis('equal')
plt.show()
```
