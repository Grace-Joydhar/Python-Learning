### **== means** 
It compares that two or more lists are equal or not. That means, is all the lists are carries same values or not?

```python
lst1 = ["jassy1", "jassy2", "jassy3"]
lst2 = ["jassy1", "jassy2", "jassy3"]

lst1 == lst2
```

**is means** not checks comparison. A list stored in a memory location and also copied in the another list. In that case, is will work.

```python
#Will not work. As both are two different lists.
lst1 = ["jassy1", "jassy2", "jassy3"]
lst2 = ["jassy1", "jassy2", "jassy3"]

lst1 is lst2
```

```python
#Will work as lst2 is containing the values of lst1.
lst1 = ["jassy1", "jassy2", "jassy3"]
lst2 = lst1

lst1 is lst2
```
