## **Iterable vs Iterators:**

```python
#Iterable: A collection of items. All the items are stored in the memory.

lst = [1,2,3,4]
for i in lst:
  print(i, end = " ")
```

**Make the iterables as iterators:** And iterate values through a iterators.

```python
make_iterator = iter(lst)
```

```python
for i in make_iterator:
  print(i, end = " ")
```

**Another way to print iterators/values:** Use next

```python
make_iterator
```

```python
next(make_iterator) #need to run the make_iterator = iter(lst) first. Then run this code
```

**To prevent the StopIteration Error, we can use try catch solution**

```python
lst1 = [1,2,3,4,5]
print("Iterables are: ")
for i in lst1:
  print(i, end = " ")
```

```python
iterate_value = iter(lst1)
```

```python
for i in iterate_value:
  print(i, end = " ")
```

```python
try:
  print(next(iterate_value))

except StopIteration:
  print("The iterator is empty")
```

## **Generators**:
It is also a useful way to create iterators. 

```python
def square(n):
  for i in range(n):
    yield i**2 #it will make the square function as iterator. But it is basically a generators. 
```

```python
for i in square(3):
  print(i)
```

```python
#Prove that Generator is also an iterator.
a = square(3)
print(a)
```

```python
next(a)
```

## **Difference between Iterators vs Generators:**
1. To create an iterator, we use iter/next function and To create generator, we use function with yeild keyword.

2. Generators saves the local variable value.

3. Generator in python helps us to write first and compact code. For iterator, it does not.

4. Python iterator is much more memory efficient.
