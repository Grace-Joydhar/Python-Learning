**Sets:** A set is an unordered collection data type that is iterable, mutable and has no duplicate values. Python set class represents the mathematical notion of a set. This is based on a data structure known as a hash table.  Represents through third bracket {}.

```python
var = set()
print(type(var))
```

```python
var = {'red', 'blue', 'black', 'red'} #will cut down the last red. As sets not allow duplicate values
print(var)
print(type(var))
```

**Indexing:** Sets does not support indexing

```python
var = {'red', 'blue', 'black', 'red'}
var['red'] #Will not Work
```

**Add values**

```python
var = {'red', 'blue', 'black'}
var.add('white') #Will add at the first position
print(var)
```

**Remove Duplicates from two different sets**

```python
set1 = {'red', 'blue', 'black', 'pink'}
set2 = {'red', 'blue', 'black', 'olive'}
set2.difference(set1) #it will remove all the same values
```

```python
set1 = {'red', 'blue', 'black', 'pink'}
set2 = {'red', 'blue', 'black', 'olive'}
set2.difference_update(set1)
print(set2)
```

**Find out similar items from the sets**

```python
set1 = {'red', 'blue', 'black', 'pink'}
set2 = {'red', 'blue', 'black', 'olive'}
set2.intersection(set1)
set2.intersection_update(set1)
print(set2)
```

**Dictionaries:** A dictionary is a collection ehich is unordered, changable and indexed. In python dictionaries are written in curly/ third bracket {} and they have keys and values.

```python
dic = {}
print(type(dic))
```

```python
dic = {1,2,3,4,5} #it is a set. because only values are present but no keys are here.
print(type(dic))
```

**Create a Dictionary**

```python
person = {'name' : 'grace', 'age': 25, 'home' : 'natore'} #it is a dict
print(type(person))

print(person['name'])
```

**Print Keys**

```python
person = {'name' : 'grace', 'age': 25, 'home' : 'natore'}
for x in person:
  print(x)
```

```python
person = {'name' : 'grace', 'age': 25, 'home' : 'natore'}
for x in person.keys():
  print(x)
```

**Print Values**

```python
person = {'name' : 'grace', 'age': 25, 'home' : 'natore'}
for x in person.values():
  print(x)
```

**Print Key and Values Together**

```python
person = {'name' : 'grace', 'age': 25, 'home' : 'natore'}
for x in person.items():
  print(x)
```

**Add a new key & value in the existing dictionary**

```python
colors = {'c1' : 'red', 'c2' : 'white'}
colors['c3'] = 'violet'
print(colors)
```

**Change a keys value** Also called overwritten

```python
colors = {'c1' : 'red', 'c2' : 'white', 'c3' : 'pink'}
colors['c2'] = 'violet'
print(colors)
```

**Nested Dictionary:**

```python
person1_age = {'Jhon' : 23}
person2_age = {'Jack' : 20}
person3_age = {'Peterx' : 20}

persons = {'person 1' : person1_age, 'person2' : person2_age, 'person3' :person3_age}
print(persons)
```

```python
print(type(persons))
```

```python
person1_age = {'Jhon' : 23}
person2_age = {'Jack' : 20}
person3_age = {'Peterx' : 20}

persons = {'person 1' : person1_age, 'person2' : person2_age, 'person3' :person3_age}
print(persons['person2'])
```

```python
tup = ()
print(type(tup))
```
