**Numpy Tutorials:**
Numpy is a general purpose array- processing package. It provides a high-performance multidimentional array object and tools for working with these arrays. It is the fundamental package for scientific computing with python.

**What is an array?**
An array is a data structure that stores values of same data type in python, this is the main difference between arrays and lists. while python lists can contain values corresponsding to different data types. arrays in python can only contain vales corresponding to same data type.

```python
import numpy as np
my_list = [1,2,3,4,5]

arr = np.array(my_list)
print(arr)
```

```python
print(type(arr))
```

```python
print(arr.shape)
```

**Create 3 lists and make an array.**

```python
list1 = [1,2,3,4,5]
list2 = [5,6,7,8,9]
list3 = [9,10,11,12,13]

arr = np.array([list1, list2, list3])
print(arr) #It will be two dimentional array

print("\nThe shape of the array is: ", arr.shape)
```

```python
result = arr.reshape(5,3)
print(result)

result1 = arr.reshape(1,15)
print("\n", result1)
```

**Indexing**

```python
arr = np.array([1,3,5,7,9,2,4])
print("All values:", arr)

print("Value of 2 number index:" , arr[2])

print(arr[4:])
```

```python
arr = np.array([[1,2,3,4,5], 
                [4,5,6,7,8], 
                [7,8,9,1,2]])
print(arr)
print("\n", arr[0:2,0:2])
print("\n", arr[1:,3:])
print("\n", arr[0: , 0:2])
print("\n", arr[1:2, 3:])
```

```python
arr = np.arange(0,10).reshape(2, 5)
print(arr)

arr1 = np.arange(0,10,2)
print("\n",arr1)
```

```python
arr = np.linspace(1,10,20, endpoint=False)
print(arr)
```

**Copy Function**

```python
arr = np.array([1,2,3,4,5])
arr[3:] = 100
print("Arr is:", arr)


#arr1 = arr #Will change the value of arr function also. Thats why we use copy function
arr1 = arr.copy()
arr1[3:] = 500
print("\nArr1 is:", arr1)

print("Arr is:", arr)
```

**Some conditions are very useful in exploratory Data Analysis.**

```python
val = 2

arr = np.array([1,2,3,4,5])
print(arr < val)
print(arr * val)
print(arr%val)
```

**Mathematics in between arrays**

```python
arr1 = np.array([1,2,3,4])
arr2 = np.array([4,3,2,1])

print("addition: ", arr1 + arr2)
print("sub: ", arr1 - arr2)
print("mul: ", arr1 * arr2)
print("div: ",arr1 / arr2)
```

**random values:**
- random.rand() : show randoms values (0-1) in between a given row and column
-random.randn(): show random values (Any) within given rows and columns
- random.random() : show any random values
- random.randint() : show any values between given values
- random.random_sample() : Will show values between 0.0 - 1.0

```python
arr = np.random.rand(4,5) #Will show randoms values (0<1) in between a given row and column
print(arr)

arr1 = np.random.randn(4,4) #Will show random values (any) within given rows and columns
print("\n", arr1)

arr2 = np.random.random(20) #Will show 20 random values
print("\n", arr2)

arr3 = np.random.randint(10,30) #Will show any values between 10 and 30
print("\n", arr3)

arr4 = np.random.randint(0,100,8).reshape(4,2) #Between 0 to 100, choose any 8 numbers every time and put in the shape of 4,2
print("\n", arr4)

arr5 = np.random.random_sample((1,5)) #Will show values between 0.0 - 1.0
print("\n", arr5)
```
