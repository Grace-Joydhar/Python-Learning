```python
class Error(Exception): #create a class and inherit a specific class
  pass

class dobException(Error): #custom exception
  pass

class addressException(Error):
  pass
```

```python
year = int(input("Enter your birth year: "))
age = 2021 - year
print("Your age is {}".format(age))

try:
  if age<=30 & age>20:
    print("You can join the exam.")
  else:
    raise dobException

except dobException:
  print("You are not able for the exam.")
```
