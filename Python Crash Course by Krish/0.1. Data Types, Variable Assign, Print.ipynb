## Ways of variable print

```python
print("My name is Grace Joydhar")
```

```python
f_name = "Grace"
l_name = "Joydhar"
print("My name is {} {}".format(f_name, l_name))
```

```python
f_name = "Grace"
l_name = "Joydhar"
print("My name is {first} {last}".format(first = f_name, last = l_name))
```
