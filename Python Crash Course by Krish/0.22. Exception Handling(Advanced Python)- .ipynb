```python
try:
  y=z

except NameError as ex1:
  print("The user have not defined the variable")

except Exception as ex: #Always need to be written at the last
  print(ex)
```

```python
try:
  a = 1
  b = "s"
  c = a + b #Typeerror. Showed in the next code

except NameError: #No alias - as ex1
  print("The user have not defined the variable")

except TypeError: #No alias - as ex2
  print("Try to make the datatype similar")

except Exception as ex: #Always need to be written at the last
  print(ex)
```

### **Try, Else Block**

```python
try:
  a = int(input("Enter Number1: "))
  b = int(input("Enter Number2: "))
  c = a/b
  d = a*b
  e = a+b

#Will print the values in else block. This is the good way to print values.
  # print("Division is: ", c)
  # print("Multiplication is: ", d)
  # print("Addition is: ", e)

except NameError: #No alias - as ex1
  print("The user have not defined the variable")

except TypeError: #No alias - as ex2
  print("Try to make the datatype similar")

except ZeroDivisionError:
  print("Please provide number greater than 0")

except Exception as ex: #Always need to be written at the last
  print(ex)

else:
  print("Division is: ", c)
  print("Multiplication is: ", d)
  print("Addition is: ", e)
```

### **Try, Else, Finally Block**

```python
try:
  a = int(input("Enter Number1: "))
  b = int(input("Enter Number2: "))
  c = a/b

#Will print the values in else block. This is the good way to print values.
  # print("Division is: ", c)

except NameError: #No alias - as ex1
  print("The user have not defined the variable")

except TypeError: #No alias - as ex2
  print("Try to make the datatype similar")

except ZeroDivisionError:
  print("Please provide number greater than 0")

except Exception as ex: #Always need to be written at the last
  print(ex)

else:
  print("Division is: ", c)

finally: #Used to close the database whatever it is working on error handling or else block.
  print("\nThe execution is done")
```
