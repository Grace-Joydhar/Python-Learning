### **Correlation with Heatmap:**
A correlation heatmap uses colored cells, typically in a monochormatic scale, to show a 2D correlation matris (table) between two discrete dimensions or event types. It is very important in Feature Selection.

- works for only integer or float values. Because Categorical features are object type. Values will be ranging between -1 to +1.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

**JoinPlot:**
A joinplot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot. 

**Bivariate Analysis:** If features are only two.

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind= 'hex') #hex means the structer will be showed as hexagonal shape
```

```python
sns.jointplot(x= 'tip', y = 'total_bill', data = df, kind= 'reg') #reg means it will draw a regression line or best fit line.
```

**Pair Plot:** A pair plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing at variables paired with all the other variables.

- If features are more than two.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

**Dis plot:** It helps us to check the distribution of the columns feature.
