```python
#Specifying columns data type
data = ('a,b,c,d\n'
          '1,2,3,4\n'
          '5,6,7,8\n'
          '9,10,11,12')
print(data)
```

```python
#with dtype, you can print the datas as you want
df = pd.read_csv(StringIO(data), dtype=object)
print(df)

df = pd.read_csv(StringIO(data), dtype=float)
print("\n",df)
```

**Specify Index column**

```python
data = ('index,a,b,c\n'
        '4,apple,bat,5,7\n'
        '8,orange,cow,10')
print(data)
```

```python
df = pd.read_csv(StringIO(data), index_col=0)
print(df)
```

```python
data = ('a,b,c\n'
        '4,apple,bat,\n'
        '8,orange,cow')
df = pd.read_csv(StringIO(data))
print(df)
```

```python
#Assigned values are not designed succesfully in the row. So use index_col = False.
df = pd.read_csv(StringIO(data),index_col=False)
print(df)
```

**Combined usecols and index_col**

```python
data = ('colors,fruits,animals\n'
        'Black,Apple,Dog,\n'
        'White,Banana,Cow')
df = pd.read_csv(StringIO(data))
print(df)
```

```python
df = pd.read_csv(StringIO(data), usecols=['colors', 'animals'], index_col=False)
print(df)
```

**Quoting and Escape characters.** *VERY USEFUL IN NLP.*

```python
data = 'a,b\n"hello, \\"Bob\\", nice to see you", How are you?'
print(data)
```

```python
df = pd.read_csv(StringIO(data), escapechar='\\')
print(df)
```
