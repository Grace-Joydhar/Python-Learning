```python
from pyforest import *
lazy_imports()
```

```python
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * x)

plt.plot(x,y, color = "maroon", linestyle = "dashed")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Demo Graph for Pyforest")
plt.show()
```

**For plotting graph, we can write all the details in one line also to reduce line numbers.**

```python
t = [10,30,50,70]
s = [20,10,50,80]

fig, ab = plt.subplots()
ab.plot(t,s, color = "black", linestyle = 'dotted', marker = "*", markersize = 10)

ab.set(xlabel='time (s)', ylabel='voltage (mV)', title='Demo Graph')
ab.grid(linewidth = 1)

# fig.savefig("test.png")
plt.show()
```
