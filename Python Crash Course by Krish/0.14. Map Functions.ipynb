### **Map Function in Python:**

**Give a list and Find out a number is even or odd**

```python
def EvenOdd(num):
  if num%2 ==0:
    return "The number {} is even".format(num) #we can write return True also
  else:
    return "The number {} is odd".format(num) #we can write return False also

lst = [1,3,2,6,44,83,56]
list(map(EvenOdd, lst)) #list(map(Function name, list name))
```
