### **Find out a number is even or odd**
**Normal Way**

```python
num = 20

if num%2 == 0:
  print("Num is Even")
else:
  print("Num is Odd")
```

**By Using Function (with user input)**

```python
num = int(input("Enter a num: "))

def even_or_odd(num):
  if num%2 == 0:
   print("Num is Even")
  else:
   print("Num is Odd")

even_or_odd(num)
```

### **Print vs Return**

**Print: ** Not store any values.

```python
def hello_world():
  print("Welcome to the world")
hello_world()
```

```python
var = hello_world()
print(var) #it will print None. Cause print function only print users input. But does not store anything
```

**Return:** Stores values.

```python
def hello_person():
  return "Nice to meet you!"
hello_person()
```

```python
value = hello_person()
print(value) #will print 'Nice to meet you!. As return always store values.
```

### **Add two numbers**

```python
num1 = int(input("Enter Num1: "))
num2 = int(input("Enter Num2: "))
 
def add_num(num1, num2):
  return (num1  * num2) + (num2 - num1)

var = add_num(num1, num2)
print("The total is: ", var)
```

### **Positional and Keyword Argument**
**- Positional Argument:** If you define a key inside a function but do not define any value for the key. That is called Positional Argument. Here name is an example of positional argument.

**- Keyword Argument:** When you will define a key value pairs inside the function, that will be worked as keyword argument. Here age = 29 is the example.

```python
def person(name, age = 25):
  print("My name is {} and my age is {}".format(name,age))
person("Jack")
```

### *args, **kwargs

```python
def hello(*args, **kwargs):
  print(args)
  print(kwargs)
hello ("Jhon", "Smith", age = 25, cell= 123456789)
```

```python
lis = ['Jhon', 'Smith']
dict_args = {'age': 25, 'cell': 123456789}

#Wrong way to print
hello(lis, dict_args)

#Right way to print with *Args, **kwargs
hello(*lis, **dict_args)
```

### **Return 2 different values**

```python
#By defining values
lst = [1,2,3,4,5,6,7]

def even_odd_sum(lst):
  even_sum = 0
  odd_sum = 0
  for i in lst:
    if i%2 == 0:
      even_sum = even_sum+i
    else:
      odd_sum = odd_sum+i
    
  return even_sum, odd_sum
var = even_odd_sum(lst)
print("Sum is:", var)
```

```python
#Through user input
lst = int(input("Enter the range of the number: "))

def even_odd_sum(lst):
  even_sum = 0
  odd_sum = 0
  for i in range(0, lst):
    if i%2 == 0:
      even_sum = even_sum+i
    else:
      odd_sum = odd_sum+i

      i = i+1
    
  return even_sum, odd_sum
var = even_odd_sum(lst)
print("Sum is:", var)
```
