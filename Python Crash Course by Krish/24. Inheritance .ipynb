Inheritance means use all the details of a class into the another class. It is kind of copying the details.

- All the class variables are public by default.
- if we use self._windows (Note: only one _), **it is protected**
- if we use self.__windows (Note: two __), **it is private**

```python
class Car():
  def __init__(self, windows, doors, enginetype):
    self.windows = windows
    self.doors = doors
    self.enginetype = enginetype
  def drive(self):
    print("Car is drives by human")
```

```python
localcar = Car(4,5, 'oil')
```

```python
localcar.doors
```

```python
localcar.drive()
```

### **Inheritance class Car** (Copying all the things from Car class.

- use super() for copying th edetails from the main class

```python
class audi(Car):
  def __init__(self, windows, doors, enginetype, enableai):
    super().__init__(windows, doors, enginetype)
    self.enableai = enableai

  def selfdriving(self):
    print("Audi car allows only Ai driving")
```

```python
AudiQ7 = audi(5, 8, 'diesel', True)
```

```python
AudiQ7.doors
```

```python
AudiQ7.enginetype
```

```python
AudiQ7.enableai
```
