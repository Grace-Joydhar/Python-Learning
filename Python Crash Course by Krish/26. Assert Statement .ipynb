### **Assert Statement :**
Python provides the assert statement to check if a given logical expression is True or False. 

Program execution proceeds only if the exception is true and raises the assertionError when it is false. The following code shows the usuage of the Assert Statement.

```python
num = 10
assert num>=10 #It is true. so it will run successfully.
```

```python
num1 = 5
assert num1 >= 130 #it is false and will raises AssertionError
```

```python
try:
  num = int(input("Enter a number: "))
  assert num%2 == 0 #It is True and it will print the given message. Otherwise ot will raises AssertionError
  print("Number is even")

except AssertionError:
  print("The number is not even number")
```
