### **Function Copy**

**- Normal way to do it with print() function.**

```python
def welcome():
  print("Welcome to our community!")

welcome()
```

```python
wel = welcome()
```

```python
wel #it will not give anything
```

**But if we use return function, put the value in a new variable, delete the previous function. Then also the output will be shown. It is called Function Copy.**

```python
def welcome():
  return "Welcome to our community!"
```

```python
wel = welcome()
del welcome
```

```python
wel
```

```python
welcome() #it will not work. as welcome() fuction has been deleted.
```

# **Closures:** 
- **Function inside a function:** Inner function can use all the variables of parent fuction.

**First way: Write the message inside the inner function**

```python
def main_wel():
  msg = "Hello friend"
  def sub_wel_class(): #it is a method
    print("Welcome to our place")
    print(msg)
    print("Please stay with us")
  return sub_wel_class()
```

```python
main_wel()
```

**Second Way: Write the message by calling the function at the last.**

```python
def main_wel(msg):
  def sub_wel_class(): #it is a method
    print("Welcome to our place")
    print(msg)
    print("Please stay with us")
  return sub_wel_class()
```

```python
main_wel("Jassy")
```

```python
def main_wel(func):
  def sub_wel_class():
    print("Hello Everyone")
    func("How are you?") #func is nothing but a print function used below
    print("Welcome to our place")
  return sub_wel_class()
```

```python
main_wel(print)
```

### **Closures and initial Decorators:**

```python
def main_wel(func):
  def sub_wel_class():
    print("Hello Everyone")
    print(func([1,2,3,4,5]))
    print("Welcome to our place")
  return sub_wel_class()
```

```python
main_wel(len) #len is a inbuilt function
```

## **Decorators:** 

```python
def main_wel(func):
  def sub_wel_class():
    print("Hello Everyone")
    func()
    print("Welcome to our place")
  return sub_wel_class()
```

```python
@main_wel #by define a decorate .This way also the initialization will happen.
def city_name():
  print("Dhaka is the capital of Bangladesh")
```

```python
#First way:
main_wel(city_name)

#Second way:
# a = main_wel(city_name)
# print(a)
```
