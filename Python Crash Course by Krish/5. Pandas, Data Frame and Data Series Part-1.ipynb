- What is dataframe? 
DataFrame is a combination of rows and columns and it is a representation of data that how the datas are look.

We read csv/excel or other formats files with data. So after reading the datas are called as a dataframe.
- What is data Science?
- Different operation in pandas.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.arange(0,20).reshape(5,4), index = ['Row1', 'Row2', 'Row3', 'Row4', 'Row5'], columns = ['Col1', 'Col2', 'Col3', 'Col4'])
print(df.head())
```

```python
df.to_csv('Text.csv')
```

**Accessing the elements**
- loc - can search the values by mentioning *row name only*. It will show all the values present on the specific row.
- iloc - can find a specific value by mentioning row and column index number.

```python
print(df[['Col3', 'Col4']])
```

```python
df.loc['Row1']
```

```python
df.iloc[2, 2]
```

```python
df.iloc[:2]
```

```python
df.iloc[0:2, 0:2]
```
