### **Categorical Plots:**
Seaborn also helps us in doing the analysis on categorical data points through
- count plot
- bar plot
- box plot
- violin plot

```python
import seaborn as sns
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')

df = pd.read_csv('/content/gdrive/My Drive/Colab_Notebooks/Sample_data/Seaborn dataset/tips.csv')
df.head()
```

**Countplot:** Show the counts of observations in each categorical bin using bars.
- Only have one axis. x or y.

```python
sns.countplot('sex', data = df)
```

**Change the axis for specific colum**

```python
sns.countplot(y = 'day', data = df)
```

**Bar Plot:** Have x and y values.

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df)
```

**Box Plot:** Values indicates percentiles

```python
sns.boxplot(x = 'sex', y = 'total_bill', data =df)
```

**Use different colors for a boxplot**

```python
sns.boxplot(x= 'day', y = 'total_bill', data = df, palette= 'rainbow')
```

```python
sns.boxplot(data = df, orient='v')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue= 'smoker', data = df)
```

**Violin Plot:** Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot.

```python
sns.violinplot(x = 'day', y ='total_bill', data = df)
```
