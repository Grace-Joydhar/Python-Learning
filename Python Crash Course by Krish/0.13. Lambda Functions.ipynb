### **Lambda Function:**
Also called as Anonymous Function. And a function with no name.
- Only can used for one expression/condition. For example, addition of two numbers or sub of multiple numbers etc.
- Faster than other functions

**Addition with normal function:**

```python
def addition(a,b):
  return a+b
result = addition(5,4)
print("Result is: ", result)
```

**Addition with Lambda Function:**
Rules: lambda variables : operation/Expression

```python
addition = lambda a,b: a+b
addition(4,6)
```

**Multiplication of 5 numbers:**

```python
#Normal Way

def multi(a,b,c,d,e):
  return a*b*c*d*e
multi(2,3,4,5,6)
```

```python
#By using Lambda Function

total = lambda a,b,c,d,e: a*b*c*d*e
total(2,3,4,5,6)
```

**Finding a number is even or not with a normal function:**

```python
#Normal Way

def evenOdd(num):
  if num%2 == 0:
    return True
  else:
    return False
evenOdd(45)
```

**Finding a number is even or not with Lambda Function:** It will automatically give us True or False.

```python
evenNot = lambda num:num%2 == 0
evenNot(456)
```
