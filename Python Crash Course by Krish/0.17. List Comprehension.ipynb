### **List Comprehension:**
List comprehension provides a concise way to create lists. It **consists of brackets** **containing an expression** followed by a **for clause, then zero or more for or if clause.** The expression can be anything, meaning you can put in all kinds of objects in lists.

**Problem:** Take a list with numbers, make squares of the numbers and store the square values in an another list.

**Normal way to solve the problem**

```python
lst1 = []

def lst_square(lst):
  for i in lst:
    lst1.append(i*i)
  return lst1

lst_square([1,2,3,4,5])
```

**With List Comprehension**

```python
lsts = [1,2,3,4,5]

new_lst = [i*i for i in lsts]
print("Square of {} are: ".format(lsts), new_lst)
```

**Problem:** Take a list of numbers from user, make squares of the numbers and store the square values in an another list.

```python
#By taking multiple values from user input at a time
nums = list(map(int, input("Enter values for doing square: ").split()))
lst1 = []

def sqr(nums):
  for i in nums:
    lst1.append(i*i)
    continue
  return lst1

sqr(nums)
print("Square of the values are: ", lst1)
```

**With List Comprehension**

```python
lst = list(map(int, input("Enter the values for doing square: ").split()))

lst1 = [i*i for i in lst] #brackets with expression then for loop/while loop/if clause
print("Square of {} are: ".format(lst), lst1)
```

**Problem:** Do square of only even numbers

```python
lst1 = []

def sqr_even(nums):
  for i in nums:
    if i%2 == 0:
      lst1.append(i*i)
  return lst1

sqr_even([1,2,3,4,5,6])
print("Squares of the even numbers are: ", lst1)
```

**With List Comprehension**

```python
lst = [1,2,3,4,5,6]

new_lst = [i*i for i in lst if i%2 == 0]
print(new_lst)
```
