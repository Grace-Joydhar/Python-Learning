Python is an object oriented programming language.

Almost everything in Python is an object, with its properties and methods.

A Class is like an object constructor, or a "blueprint" for creating objects.

For example, class as a sketch (prototype) of a house. It contains all the details about the floors, doors, windows etc. Based on these descriptions we build the house. House is the object.

# **Create Class**
To create a class, use the keyword class

```python
class myclass:
  x = 10
print(myclass)
```

# **Create Object**
Now we can use the class named MyClass to create objects

```python
#Sketch of a house contains all the details about the house.
class myclass:
  x = 5
  y = 10
  z = x*y

#Build the house including all the details from the sketch (Class)
obj = myclass()
print(obj.z)
```

# **The __init__() Function**
The examples above are classes and objects in their simplest form, and are not really useful in real life applications.

To understand the meaning of classes we have to understand the built-in __init__() function.

All classes have a function called __init__(), which is always executed when the class is being initiated.

Use the __init__() function to assign values to object properties, or other operations that are necessary to do when the object is being created:

```python
#Sketch of a person
class person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

#The person
p1 = person('Jason', 25)
p2 = person('Jessi', 40)

print ('My friend\'s name is', p1.name, 'and his age is', p1.age)
print ('His mom\'s name is', p2.name, 'and her age is', p2.age)
```

# **Object Methods**
Objects can also contain methods. Methods in objects are functions that belong to the object.

```python
#Insert a function that prints a greeting, and execute it on the p1 object:

class person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def opening(self):
    print("Welcom to my world")

  def closing(self):
    print("Good Bye")

p1 = person('Jessi', 26)

p1.opening()

print('My name is', p1.name, 'and my age is', p1.age)

p1.closing()
```

# **Modify Object Properties**

```python
class person():
  def __init__(self, name, age, year):
    self.name = name
    self.age = age
    self.year = year

p1 = person('John', 25, 2000)

print('In', p1.year, p1.name, 'was', p1.age, 'years old.')

p1.age = 45

print('But now he is', p1.age,'.')

```

# **Delete Object Properties:**
Using del()

```python
class person():
  def __init__(self, name, age, year):
    self.name = name
    self.age = age
    self.year = year

p1 = person('John', 25, 2000)
print(p1.name, 'is', p1.age, 'years old in', p1.year)

del p1.year
print(p1.name, 'is', p1.age, 'years old in', p1.year)
```

# **The pass Statement**

```python
class person():
  def __init__(self, name, age, year):
    self.name = name
    self.age = age
    self.year = year

  def greeting(self):
    pass

p1 = person('John', 25, 2000)
print(p1.name, 'is', p1.age, 'years old in', p1.year)
```
