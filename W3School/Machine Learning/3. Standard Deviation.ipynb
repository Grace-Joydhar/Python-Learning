### **What is Standard Deviation?**
Standard deviation is a number that describes how spread out the values are.

**A low standard deviation** means that most of the numbers are close to the mean (average) value.

**A high standard deviation** means that the values are spread out over a wider range.

**Example:** This time we have registered the speed of 7 cars within a similar range:

speed = [86,87,88,86,87,85,86]

What will be the standard deviation? ***It will be low standard deviation.***

```python
#Find the mean and standard deviation

import numpy as np

speed = [86,87,88,86,87,85,86]

mean= np.mean(speed)
print("Mean is: ", mean)

sd = np.std(speed)
print("standard deviation is:", sd)
```

Let us do the same with a selection of numbers with a wider range:

speed = [32,111,138,28,59,77,97]

What will be the standard deviation? ***It will be high standard deviation.***

```python
#Find mean and standard deviation
import numpy as np

speed = [32,111,138,28,59,77,97]

mean = np.mean(speed)
print("Mean is: ", mean)

sd = np.std(speed)
print("Standard deviation is: ", sd)
```

### **Variance**
Variance is another number that indicates how spread out the values are.

In fact, if you take the square root of the variance, you get the standard deviation!

Or the other way around, if you multiply the standard deviation by itself, you get the variance!

To calculate the variance you have to do as follows:
1. Find the mean
2. For each value: find the difference from the mean
3. For each difference: find the square value

```python
#Find Variance
#From variance, find standard deviation

import numpy as np

num = [32,111,138,28,59,77,97]
sort = np.sort(num)
print("Sorting in ascending order: ", sort)

#Variance
var = np.var(num)
print(var)

#Standard Deviation
sd = np.sqrt(var)
print("Standard Deviation is:", sd)

#or as previous, we can directly find the standard deviation from the mentioned numbers:
sd1 = np.std(num)
print("Standard deviation by using the normal way:" , sd1)
```
