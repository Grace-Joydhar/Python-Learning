### **Mean, Median, and Mode**
In Machine Learning (and in mathematics) there are often three values that interests us:

**1. Mean** - The average value
**2. Median** - The mid point value
**3. Mode** - The most common value

**Example:** We have registered the speed of 13 cars:

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

**What is the average, the middle, or the most common speed value?**

```python
#Sorting numbers
import numpy as np

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#Ascending Order
result = np.sort(speed) 
print("Ascending Order:", result)

#Descending Order
result[::-1].sort()
print("Descending Order:", result)
```

```python
#Mean(Avg):
import numpy as np

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

mean = np.mean(speed)
print(mean)
```

```python
#Median(Mid Point):
import numpy as np

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

median = np.median(speed)
print(median)
```

Numpy has no attribute named mode. So we have to import stats from scipy.

*The mode() method returns a ModeResult object that contains the mode number (86), and count (how many times the mode number appeared (3)).*

```python
#Mode(Most Common Value):
import scipy
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

mode = stats.mode(speed)
print(mode)
```
