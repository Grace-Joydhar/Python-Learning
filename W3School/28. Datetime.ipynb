### **Python Dates**
A date in Python is not a data type of its own, but we can import a module named datetime to work with dates as date objects.

**Import the datetime module and display the current date**

**Output Explaination:** *Year-Month-Date Hours:Minutes:second:Micro Seconds *

```python
import datetime

present_time = datetime.datetime.now()
print(present_time)
```

## **Notes for Datetime**
**Directive  	  Description	    Example**	 
**%a**	Weekday, short version -	Wed	
**%A**	Weekday, full version	- Wednesday	
**%w**	Weekday as a number 0-6 - 0 is Sunday	3	
**%d**	Day of month 01-31 - 31	
**%b**	Month name, short version	- Dec	
**%B**	Month name, full version -	December	
**%m**	Month as a number - 01-12	12	
**%y**	Year, short version, without century -	18	
**%Y**	Year, full version -	2018	
**%j**	Day number of year 001-366 -	365	
**%U**	Week number of year, Sunday as the first day of week, - 00-53	52	
**%W**	Week number of year, Monday as the first day of week, - 00-53	52	


**%H**	Hour 00-23	- 17	
**%I**	Hour 00-12	- 05		
**%M**	Minute 00-59	- 41	
**%S**	Second 00-59 -	08	
**%f**	Microsecond - 000000-999999	548513	
**%z**	UTC offset	- +0100	
**%Z**	Timezone	- CST	

**%c**	Local version of date and time	- Mon Dec 31 17:41:00 2018	
**%p**	AM/PM -	PM
**%x**	Local version of date	- 12/31/18	
**%X**	Local version of time -	17:41:00	

**%%**	A % character	- %	
**%G**	ISO 8601 year	- 2018	
**%u**	ISO 8601 weekday (1-7) -	1	
**%V**	ISO 8601 weeknumber (01-53)	- 01

### **For the present Time**
1. Return the year %y or %Y
2. name of weekday %a or %A
3. number of the week of the month %w
4. number of the week of the year %W
5. Month Number %m
6. Month name %b or %B
7. Day of the month %d
8. Day of the year %j
9. Present time %c, AM/ PM - %p
10. Local verson of date and time %x and %X

```python
import datetime

present_time = datetime.datetime.now()

print("This is the year of", present_time.strftime("%Y"))
print("Now the weekday is", present_time.strftime("%A")) 
print("Week Number of the month is:", present_time.strftime("%w"))
print("Week Number of the year is:", present_time.strftime("%W"))
#%U means	Week number of year, Sunday as the first day of week, - 00-53	52 
#%W	means Week number of year, Monday as the first day of week, - 00-53	52
print("This is the", present_time.strftime("%m"), "month of the year")
print("The month name is", present_time.strftime("%B"))
print("Today is the", present_time.strftime("%d"), "number of day of the month")
print("Today is the", present_time.strftime("%j"), "number of day of the year")
print("Now the time is:", present_time.strftime("%c"))
print("Local version of time is:", present_time.strftime("%X"), present_time.strftime("%p")) #%p means AM/PM
print("Local version of date is:", present_time.strftime("%x"))
```

### **For any time of the year**
1. Return the year %y or %Y
2. name of weekday %a or %A
3. number of the week of the month %w
4. number of the week of the year %W
5. Month Number %m
6. Month name %b or %B
7. Day of the month %d
8. Day of the year %j
9. Local verson of date %x

```python
import datetime

present_time = datetime.datetime(2014, 10, 11)

print("This is the year of", present_time.strftime("%Y"))
print("The weekday is", present_time.strftime("%A")) 
print("Week Number of the month is:", present_time.strftime("%w"))
print("Week Number of the year is:", present_time.strftime("%W"))
#%U means	Week number of year, Sunday as the first day of week, - 00-53	52 
#%W	means Week number of year, Monday as the first day of week, - 00-53	52
print("This is the", present_time.strftime("%m"), "month of the year")
print("The month name is", present_time.strftime("%B"))
print("Today is the", present_time.strftime("%d"), "number of day of the month")
print("Today is the", present_time.strftime("%j"), "number of day of the year")
print("Local version of date is:", present_time.strftime("%x"))
```
