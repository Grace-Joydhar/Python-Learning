JSON is a syntax for storing and exchanging data.

JSON is text, written with JavaScript object notation.

**Parse JSON** - Convert from JSON to Python. It will provide a python dictionary.
- json.loads()

```python
import json

details = '{"name": "Smith", "age": 30, "city": "BD"}' #json

convert = json.loads(details)

print(convert)
```

**Convert from Python to JSON**
If you have a Python object, you can convert it into a JSON string by using the json.dumps() method.

```python
import json

details = {"name": "Smith", "age": 30, "city": "BD"} #Python dictionary

convert = json.dumps(details)

print(convert)
```

**You can convert Python objects of the following types, into JSON strings:**

dict, list, tuple, string, int, float, True, False, None

**Python	JSON**
1. dict -	Object
2. list -Array
3. tuple	- Array
4. str	- String
5. int	- Number
6. float	- Number
7. True	- true
8. False	- false
9. None	- null

```python
import json

print(json.dumps({"name": "smith", "Age": 30, "city": "NY"})) #Dictionary
print(json.dumps(("apple", "banana", "orange"))) #List
print(json.dumps(["red", "green", "blue"])) #Tuple
print(json.dumps("Bangladesh")) #String
print(json.dumps(30)) #integer
print(json.dumps(12.56)) #float
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

**Convert a Python object containing all the legal data types**

```python
import json

x = {"name":"jhon",
     "age" : 30,
     "married": True,
     "divorced": False,
     "children": ("Ann", "Jerry"),
     "pets": None,
     "cars": [{"model": "BMW 230", "mpg": 27.5}, {"model": "Ford Edge", "mpg": 24.1}]
     }

y = json.dumps(x)

print(y)
```

**Format the Result**
The example above prints a JSON string, but it is not very easy to read, with no indentations and line breaks.

The json.dumps() method has parameters to make it easier to read the result:
1. indent = ex:4
2. separators=(". ", " = ")
3. sort_keys=True

```python
import json

x = {"name":"jhon",
     "age" : 30,
     "married": True,
     "divorced": False,
     "children": ("Ann", "Jerry"),
     "pets": None,
     "cars": [{"model": "BMW 230", "mpg": 27.5}, {"model": "Ford Edge", "mpg": 24.1}]
     }

y = json.dumps(x, indent = 3, separators=(".","="), sort_keys= True) #indent for making space, separrators for separate items from another, sort_keys means print all the items in accending order.

print(y)
```
