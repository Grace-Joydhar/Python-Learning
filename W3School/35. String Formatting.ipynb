To make sure a string will display as expected, we can format the result with the format() method.

### **String format()**
The format() method allows you to format selected parts of a string.
Add placeholders (curly brackets {}) in the text, and run the values through the format() method:

```python
price = 49
txt = "I want to buy the food with {} dollars"
print(txt.format(price))
```

Format the price to be displayed as a number with two decimals:

```python
price = 49
txt = "I want to buy the food with {:.2f} dollars"
print(txt.format(price))
```

### **Multiple Values**
If you want to use more values, just add more values to the format() method:

```python
item = "apples"
amount = 35
price = 150
txt = "I want to buy {} kg {} by {} taka" 
print(txt.format(amount, item, price)) #have to write the variables based on the txt sequence.
```

### **Named Indexes**
You can also use named indexes by entering a name inside the curly brackets {item}, but then you must use names when you pass the parameter values txt.format(item = "apples"):

```python
item = "apples"
amount = 35
price = 150
txt = "I want to buy {amount} kg {item} by {price} taka" 
print(txt.format(item = "apples", amount = 35, price = 150))
```

### **Index Numbers**
You can use index numbers (a number inside the curly brackets {0}) to be sure the values are placed in the correct placeholders:

```python
item = "apples"
amount = 35
price = 150
txt = "I want to buy {1} kg {0} by {2} taka" 
print(txt.format(item, amount, price))
```

Repetition of Index Numbers: We can use same index number in many places.

```python
item = "apples"
amount = 45
price = 130
txt = "I want to buy {1} kg {0}. \nThe price of {0} should be {2} dollars"
print(txt.format(item, amount, price))
```

### **The Placeholders**
The placeholders can be identified using named indexes {price}, numbered indexes {0}, or even empty placeholders {}.

```python
txt1 = "My name is {name} and I am {age} years old".format(name = "Smith", age = 30)
txt2 = "My name is {name} and I am {age} years old".format(name = "Ann", age = 34)

print(txt1)
print(txt2)
```
